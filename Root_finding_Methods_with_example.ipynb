## NEWTONS METHOD 

```python
''''THIS CODE PERFORMES SECANT'S METHOD calculate the root of the function f(x)'''
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
# from sympy import *
# x=symbols('x')
f1=lambda x: x+np.e**(-x**2)*np.cos(x)
df1=lambda x:1 - 2.0*np.e**(-x**2)*x*np.cos(x) - np.e**(-x**2)*np.sin(x)
ddf1=lambda x: np.e**(-x**2)*(4.0*x**2*np.cos(x) + 4.0*x*np.sin(x) - 3.0*np.cos(x))
f2=lambda x:(x+np.e**(-x**2)*np.cos(x))**2
df2=lambda x: (x + np.e**(-x**2)*np.cos(x))*(2 - 4.0*np.e**(-x**2)*x*np.cos(x) - 2*np.e**(-x**2)*np.sin(x))
ddf2=lambda x:(-2 + 4.0*np.e**(-x**2)*x*np.cos(x) + 2*np.e**(-x**2)*np.sin(x))*(-1 + 2.0*np.e**(-x**2)*x*np.cos(x) + np.e**(-x**2)*np.sin(x)) + np.e**(-x**2)*(x + np.e**(-x**2)*np.cos(x))*(8.0*x**2*np.cos(x) + 8.0*x*np.sin(x) - 6.0*np.cos(x))
# print(f1_x,df1_x)
tol=10**(-6)
# print(tol)
def Newtons(P0,fun,tol,der):
    Pg=P0
    f_xit=[]
    f_xit.append(fun(Pg))
    df_x=der(Pg)
    cnt=0
    itr=[]
    itr.append(np.array([cnt]))
    x_val=[]
    x_val.append(Pg)
    while True:
        Pt=Pg
        Pg=Pt-(fun(Pt))/(der(Pt))
#         print(Pg)
        f_xit.append(fun(Pg))
        x_val.append(Pg)
        error =np.array(np.abs((Pg-Pt)/Pg),dtype=np.float64)
#         print(error)
        cnt+=1
        itr.append(np.array([cnt]))
#         print(cnt)
        if (error<tol):
            break;
    return f_xit,itr,x_val
fval1,niter1,xval1=Newtons(0,f1,tol,df1)
fval2,niter2,xval2=Newtons(0,f2,tol,df2)
# q=[]
# for k in range(2,len(xval1)-1):
#     q.append(log(np.abs((xval2[k+1]-xval2[k])/(xval2[k]-xval2[k-1])))/log(np.abs((xval2[k]-xval2[k-1])/(xval2[k-1]-xval2[k-2]))))
# print("order of convergence= {}".format(np.mean(q)))
print("the root value of function(1) : {}".format(xval1[-1]))
print("the root value of function(2) : {}".format(xval2[-1]))
print("number of iterations for f1 using Newtons Method = {}".format(niter1[-1]))
print("number of iterations for f2 using Newtons Method = {}".format(niter2[-1]))
plt.style.use('fivethirtyeight')
plt.figure(1,figsize = [10,5])
plt.suptitle("Approximation using Newtons Method")
plt.subplot(121)
plt.title("f1(x) vs  x")
plt.ylabel('f(x)')
plt.xlabel('x')
plt.plot(xval1,fval1,'bo')
plt.subplot(122)
plt.title("f2(x) vs  x")
plt.xlabel('x')
plt.plot(xval2,fval2,'ro')
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.8, 
                    wspace=0.8, 
                    hspace=0.6)
dictionary = {'iteration': niter1, 'x-val': xval1, 'f(x)': fval1}  
dataframe = pd.DataFrame(dictionary) 
dataframe.to_csv('newton1.csv')
pnt1=pd.read_csv('newton1.csv',index_col=0)
print("the detail of newtons methon on function 1 :\n")
print(pnt1)
dictionary = {'iteration': niter2, 'x-val': xval2, 'f(x)': fval2}  
dataframe = pd.DataFrame(dictionary) 
dataframe.to_csv('newton2.csv')
pnt2=pd.read_csv('newton2.csv',index_col=0)
print("the detail of newtons methon on function 2 :\n")
pnt2
        
    
    
```

## SECANT METHOD

```python
'''THIS CODE PERFORMES SECANT'S METHOD'''
def Secant(P0,P1,fun,tol):
    Pg0=P0
    Pg1=P1
    q=fun(Pg1)
    p=fun(Pg0)
    Pg=Pg1-(((fun(Pg1))*(Pg1-Pg0))/(q-p))
    f_xit=[]
    f_xit.insert(0,p)
    f_xit.insert(1,q)
    f_xit.append(fun(Pg))
    cnt=0
    itr=[]
    itr.insert(0,0)
    itr.insert(1,0)
    x_val=[]
    x_val.insert(0,Pg0)
    x_val.insert(1,Pg1)
    x_val.append(Pg)
    itr.append(np.array([cnt]))
#     print(x_val)
    while True:
        Pt0=x_val[-2]
        Pt1=x_val[-1]
        Pg=Pt1-(((fun(Pt1))*(Pt1-Pt0))/(fun(Pt1)-fun(Pt0)))
#         print(Pg)
        f_xit.append(fun(Pg))
        x_val.append(Pg)
        error =np.array(np.abs((Pg-Pt1)/Pg),dtype=np.float64)
#         print(error)
        cnt+=1
        itr.append(np.array([cnt]))
#         print(cnt)
        if (error<tol):
            break;
    return f_xit,itr,x_val
fval1,niter1,xval1=Secant(0,1,f1,tol)
fval2,niter2,xval2=Secant(0,1,f2,tol)
print("the root value of function(1) : {}".format(xval1[-1]))
print("the root value of function(2) : {}".format(xval2[-1]))
print("number of iterations for f1 using Secant Method = {}".format(niter1[-1]))
print("number of iterations for f2 using Secant Method = {}".format(niter2[-1]))
plt.style.use('fivethirtyeight')
plt.figure(1,figsize = [10,5])
plt.suptitle("Approximation using Secant Method")
plt.subplot(121)
plt.title("f1(x) vs  x")
plt.ylabel('f(x)')
plt.xlabel('x')
plt.plot(xval1,fval1,'bo')
plt.subplot(122)
plt.title("f2(x) vs  x")
plt.xlabel('x')
plt.plot(xval2,fval2,'ro')
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.8, 
                    wspace=0.8, 
                    hspace=0.6)
# print(len(niter1),len(fval1),len(xval1))
dictionary = {'iteration': niter1[1:-1], 'x-val': xval1[1:-1], 'f(x)': fval1[1:-1]}  
dataframe = pd.DataFrame(dictionary) 
dataframe.to_csv('secant1.csv')
pnt1=pd.read_csv('secant1.csv',index_col=0)
print("the detail of newtons methon on function 1 :\n")
print(pnt1)
dictionary = {'iteration': niter2[1:-1], 'x-val': xval2[1:-1], 'f(x)': fval2[1:-1]}  
dataframe = pd.DataFrame(dictionary) 
dataframe.to_csv('secant2.csv')
pnt2=pd.read_csv('secant2.csv',index_col=0)
print("the detail of newtons methon on function 2 :\n")
pnt2
```

## MODIFIED NEWTONS METHOD

```python
'''THIS CODE PERFORMES MODIFIED NEWTONS METHOD'''
def Mod_Newtons(P0,fun,tol,df,ddf):
    Pg=P0
    f_xit=[]
    f_xit.append(fun(Pg))
    cnt=0
    itr=[]
    itr.append(np.array([cnt]))
    x_val=[]
    x_val.append(Pg)
    while True:
        Pt=Pg
        p=fun(Pt)
        q=df(Pt)
        m=ddf(Pt)
        Pg=Pt-(p*q)/(q**2-p*m)
#         print(Pg)
        f_xit.append(fun(Pg))
        x_val.append(Pg)
        error =np.array(np.abs((Pg-Pt)/Pg),dtype=np.float64)
#         print(error)
        cnt+=1
        itr.append(np.array([cnt]))
#         print(cnt)
        if (error<tol):
            break;
    return f_xit,itr,x_val
fval1,niter1,xval1=Mod_Newtons(0,f1,tol,df1,ddf1)
fval2,niter2,xval2=Mod_Newtons(0,f2,tol,df2,ddf2)
print("the root value of function(1) : {}".format(xval1[-1]))
print("the root value of function(2) : {}".format(xval2[-1]))
print("number of iterations for f1 using Modified Newtons Method = {}".format(niter1[-1]))
print("number of iterations for f2 using Modified Newtons Method = {}".format(niter2[-1]))
plt.style.use('fivethirtyeight')
plt.figure(1,figsize = [10,5])
plt.suptitle("Approximation using Modified Newtons Method")
plt.subplot(121)
plt.title("f1(x) vs  x")
plt.ylabel('f(x)')
plt.xlabel('x')
plt.plot(xval1,fval1,'bo')
plt.subplot(122)
plt.title("f2(x) vs  x")
plt.xlabel('x')
plt.plot(xval2,fval2,'ro')
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.8, 
                    wspace=0.8, 
                    hspace=0.6)
dictionary = {'iteration': niter1, 'x-val': xval1, 'f(x)': fval1}  
dataframe = pd.DataFrame(dictionary) 
dataframe.to_csv('mdn1.csv')
pnt1=pd.read_csv('mdn1.csv',index_col=0)
print("the detail of newtons methon on function 1 :\n")
print(pnt1)
dictionary = {'iteration': niter2, 'x-val': xval2, 'f(x)': fval2}  
dataframe = pd.DataFrame(dictionary) 
dataframe.to_csv('mdn2.csv')
pnt2=pd.read_csv('mdn2.csv',index_col=0)
print("the detail of newtons methon on function 2 :\n")
pnt2
```

```python
'''This code performed the iterative method to calculate the value of roots after taking initial guess in radians'''
import numpy as np
import matplotlib.pyplot as plt
r1=10
r2=6
r3=8
r4=4
#TAKING gUESS VALUES IN RADIAN
f=lambda θ2,θ3 :np.array([[r2*np.cos((θ2))+r3*np.cos((θ3))+r4*np.cos((220*np.pi/180))-r1],[r2*np.sin((θ2))+r3*np.sin((θ3))+r4*np.sin((220*np.pi/180))]])
df=lambda θ2,θ3:np.array([[-r2*np.sin((θ2)),-r3*np.sin((θ3))],[r2*np.cos((θ2)),r3*np.cos((θ3))]])
# f=lambda θ2,θ3 :np.array([[r2*np.cos(np.radians(θ2))+r3*np.cos(np.radians(θ3))+r4*np.cos(np.radians(220))-r1],[r2*np.sin(np.radians(θ2))+r3*np.sin(np.radians(θ3))+r4*np.sin(np.radians(220))]])
# df=lambda θ2,θ3:np.array([[-r2*np.sin(np.radians(θ2)),-r3*np.sin(np.radians(θ3))],[r2*np.cos(np.radians(θ2)),r3*np.cos(np.radians(θ3))]])
# J=df(θ0[0].item(),θ0[1].item())
# print(J)
# print(np.linalg.inv(J))
def Non_linear(θ0,tol):
    θg=θ0
    θ_val=[]
    f_val=[]
    θ_val.append(θg)
    f_val.append(f(θg[0].item(),θg[1].item()))
    cnt=0
    itr=[]
    itr.append(np.array([cnt]))
    ferr=[]
    while True:
        θs=θg
        θg=θs-np.matmul(np.linalg.inv(df(θs[0].item(),θs[1].item())),f(θs[0].item(),θs[1].item()))
        θ_val.append(θg)
        f_val.append(f(θg[0].item(),θg[1].item()))
        error=np.abs(np.abs(θg-θs)/θg)
        ferr.append(np.abs(f(θg[0].item(),θg[1].item())-f(θs[0].item(),θs[1].item())))
        cnt+=1
        itr.append(np.array([cnt]))
        if (error.all()<tol):
            break
    return θ_val,f_val,itr,ferr
# θ0=np.array([[30],[0]])
θ0=np.array([[30*np.pi/180],[0]])
tol=10**(-6)
tv,fv,itrn,ferr=Non_linear(θ0,tol)
print("--------------TAKING INITAIAL GUESS IN RADIANS-----------------------")
print("the root theeta  value of the non linear system : {}".format(tv[-1]*180/np.pi))
print("the function value of the non linear system : {}".format(fv[-1]))
print("the nuber of iteration: {}".format(itrn[-1]))

        
        
                        
```
